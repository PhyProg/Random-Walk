```python
from env.light_grid import LightGrid
from walker.light_walker import LightWalker
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def processor(coord):
    return np.sqrt(coord[0] ** 2 + coord[1] ** 2)
```

```python
grid = LightGrid(bounded = True, dim = 2, bounds = [None, None])
grid.add_processor(processor, 'Fon')
```

```python
#Wormholes

dim = 2
no_of_wormholes = 2000

for i in range(no_of_wormholes):
    first = np.random.randint(low = -50, high = 50, size = dim)
    second = np.random.randint(low = -50, high = 50, size = dim)
    grid.add_wormhole(first, second, one_way = False)
```

```python
#grid.wormholes
```

```python
collection_key = 'Fon'
```

```python
walker = LightWalker(data_collection_keys = [collection_key], dim = 2)
```

```python
#walker.__dict__
```

```python
#grid.__dict__
```

```python
no_of_steps = int(1e4)
```

```python
no_of_steps
```

```python
w = 0
```

```python
skip_next = False
for i in range(no_of_steps):
    curr = walker.get_current_position()
    walker.add_data(collection_key, grid.processors[collection_key](curr))
    next_positions = grid._next(curr, skip_next)
    if skip_next:
        skip_next = False
    energy = []
    for pos in next_positions:
        energy.append(grid.processors[collection_key](pos))
    walker.walk(next_positions, energy = energy)
    if str(walker.get_current_position()) in grid.wormholes.keys() and str(walker.get_prev_position()) in grid.wormholes.keys():
        #print(walker.get_current_position())
        skip_next = True
        w += 1
```

```python
traj = walker.get_visited_coordinates()
```

```python
x, y = traj[:, 0], traj[:, 1]
```

```python
time = np.linspace(1, no_of_steps, no_of_steps)
```

```python
fig, ax = plt.subplots()

fig.set_figwidth(15)

ax.plot(x, label = 'X')
ax.plot(y, label = 'Y')
ax.plot(time, walker.data[collection_key], label = 'distance')

ax.grid()

ax.legend()
```
